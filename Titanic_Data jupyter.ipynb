```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\ASUS\Downloads\titanic_data.xlsx")
df
```

```python
# Basic Data Overview
df.info()
```

```python
# Statistical Summary
df.describe(include='all')
```

```python
# Check for Missing Values
df.isnull().sum()
```

```python
# Count of unique values per column
df.nunique()
```

```python
# Preview last 5 rows (optional)
df.tail()
```

```python
# Data Cleaning

# Fill missing 'Embarked' with mode (recommended syntax to avoid chained assignment warning)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Fill missing 'Age' with median (safe syntax)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Drop 'Cabin' column due to too many missing values
df.drop(columns='Cabin', inplace=True)
```

```python
# Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True, color = 'blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Fare by Passenger Class
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Pclass', y='Fare', color = 'Indigo')
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
#  Countplot of Survival by Sex
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Survived', hue='Sex')
plt.title('Survival by Sex')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Age vs Fare colored by Survival
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'SibSp']], hue='Survived', diag_kind='kde')
plt.figure(figsize=(8, 5))
plt.suptitle('Pairplot of sisb by Survival', y=1.02)
plt.show()

```

```python
# Survival by Pclass (Passenger Class)
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Set style
sns.set(style="whitegrid")

# Bar plot of Survived
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, color='DarkCyan')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```
